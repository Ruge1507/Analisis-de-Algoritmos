# **Introducción**:

En el área de la ciencia de datos no es raro encontrarse con bases de datos con miles o, incluso, millones de registros que, a su vez, tienen múltiples variables que los definen. Es por eso que en esta rama de las ciencias no basta con conocer un lenguaje de programación y tener la habilidad de elaborar algoritmos en él, sino que es necesario ir más allá de lograr el objetivo deseado y analizar el uso de recursos del algoritmo creado.

Es aquí donde entra el análisis de algoritmos, que no solo se enfoca en la eficacia de lo programado, sino también en su eficiencia. Es muy importante hacer el análisis de la cantidad de recursos que utilizará un algoritmo, tanto en tiempo como en memoria, pues dependiendo del orden de crecimiento que pueda tener un programa, el simple cambio de una unidad en el tamaño de la entrada del algoritmo puede implicar un incremento exorbitante en el tiempo de ejecución.

El próposito de este reporte es justamente hacer una comparación de los posibles órdenes de crecimiento en la cantidad de tiempo utilizada por un algoritmo. Con base en la notación «Big O» se realizarán cinco comparaciones gráficas, así como numéricas, de órdenes de crecimiento que, como se menciona en [1], son de las más conocidas y, hasta cierto punto, de las más comunes; estas son las órdenes de crecimiento polinomiales, logarítmicas, factoriales y exponenciales.

# **Gráficos de Comparación**:

```python
#Importamos las librerías necesarias para llevar a cabo la práctica
import matplotlib.pyplot as plt
import math
import pandas as pd
```

### **$O$$(1)$ vs $O$($log_2$ ($n$))**

En este caso procederemos a comparar un orden de crecimiento constante contra un orden de crecimiento logaítmico, para lo cual usaremos un rango de graficación de entradas de tamaño 1 hasta 1000.

```python
#Creamos los datos a graficar estableciendo los valores de tamaño de entrada y definiendo las funciones dentro de las listas.
n = list(range(1,1001))
O1 = [1 for i in n]
Olog = [math.log2(i) for i in n]

#Graficamos los resultados obtenidos.
plt.plot(n, O1, label='y = 1', color='blue', linewidth=2)
plt.plot(n, Olog, label='y = $log_2(n)$', color='red', linewidth=2)
plt.title('Comparación de orden de crecimiento O(1) vs O($log_2(n)$)')
plt.xlabel('Tamaño de la entrada n')
plt.ylabel('Orden de crecimiento')
plt.grid(True)
plt.legend()
plt.show()
```

Observaciones del gráfico: En este caso es posible graficar utilizando el rango completo propuesto; es decir, tamaños de entrada desde 1 hasta 1000, ya que tanto el orden constante como el orden logarítmico son de los más ideales para el tiempo de ejecución de un algoritmo, aunque poco frecuentes en la práctica, especialmente el de orden constante, pues un algoritmo recibe una entrada para  producir una salida, por lo que generalmente es necesario procesar dicha entrada. Aunque el orden de crecimiento logarítmico no es constante, vemos que su incremento no es relativamente grande, por lo que el tiempo de ejecución no se ve significativamente alterado por la modificación en el tamaño de la entrada.

### **$O$($n$) vs $O$($nlog_2$($n$))**

Ahora procedemos a comparar un orden de crecimiento lineal contra un orden de crecimiento $nlog_2(n)$, para lo cual usaremos un rango de graficación de entradas de tamaño 1 hasta 1000, pues estos órdenes aún se consideran bastante aceptables.

```python
#Creamos los datos a graficar estableciendo los valores de tamaño de entrada y definiendo las funciones dentro de las listas.
n = list(range(1,1001))
On = [i for i in n]
Onlog = [i*math.log2(i) for i in n]

#Graficamos los resultados obtenidos.
plt.plot(n, On, label='y = $n$', color='blue', linewidth=2)
plt.plot(n, Onlog, label='y = $nlog_2(n)$', color='red', linewidth=2)
plt.title('Comparación de orden de crecimiento O($n$) vs O($nlog_2(n)$)')
plt.xlabel('Tamaño de la entrada n')
plt.ylabel('Orden de crecimiento')
plt.grid(True)
plt.legend()
plt.show()
```

Observaciones del gráfico: Para esta comparación también es posible contemplar el rango completo propuesto de tamaños de entrada $n$, ya que, si bien se nota un aumento en la escala del gráfico, los tiempos de ejecución aún son considerablemente buenos. Vemos que la diferencia entre ambos órdenes ya es mucho mayor que la diferencia entre los dos previos.  

### **$O$($n^2$) vs $O$($n^3$)**

A continuación, se  realiza la comparación entre órdenes polinomiales, específicamente entre el crecimiento cuadrático y el crecimiento cúbico. A partir de estos órdenes, comenzamos a ver que los tiempos de ejecución sí incrementan considerablemente cuando se aumenta el tamaño de la entrada. Por tal motivo, se decide usar un rango de graficación reducido, específicamente tamaños de entrada desde 1 hasta 200.

```python
#Creamos los datos a graficar estableciendo los valores de tamaño de entrada y definiendo las funciones dentro de las listas.
n = list(range(1,201))
On2 = [i**2 for i in n]
On3 = [i**3 for i in n]

#Graficamos los resultados obtenidos.
plt.plot(n, On2, label='y = $n^2$', color='blue', linewidth=2)
plt.plot(n, On3, label='y = $n^3$', color='red', linewidth=2)
plt.title('Comparación de orden de crecimiento O($n^2$) vs O($n^3$)')
plt.xlabel('Tamaño de la entrada n')
plt.ylabel('Orden de crecimiento')
plt.grid(True)
plt.legend()
plt.show()
```

Observaciones del gráfico: Entramos en el área de órdenes de incremento en el que las variaciones del tamaño de entrada $n$ provocan modificaciones considerables en los tiempos de ejecución. En esta gráfica que comparamos el crecimiento cuadrático contra el crecimiento cúbico, vemos que este último incrementa tanto que hace parecer al cuadrático como si fuera casi constante debido a la escala tan diferente en la que cada uno se desarrolla. Por este motivo solo se tomaron entradas de tamaño máximo 200 para graficar.

### **$O$($a^n$) vs $O$($n!$)**

Continuamos con una comparación entre el crecimiento exponencial, donde usaremos $a=2$, y el crecimiento factorial. Como se menciona en [2], los órdenes exponenciales y factoriales ya son considerados órdenes de muy alto costo, pues en estos órdenes el incremento en pocas unidades del tamaño de entrada del algoritmo ya representa un aumento verdaderemente grande en el tiempo de ejecución. Este es el motivo por el que se decide utilizar como rango de graficación únicamente tamaños de entrada desde 1 hasta 10.

```python
#Creamos los datos a graficar estableciendo los valores de tamaño de entrada y definiendo las funciones dentro de las listas.
n = list(range(1,11))
O2n = [2**i for i in n]
O3n = [3**i for i in n]
O5n = [5**i for i in n]
Ofactn = [math.factorial(i) for i in n]

#Graficamos los resultados obtenidos.
plt.plot(n, O2n, label='y = $2^n$', color='blue', linewidth=2)
plt.plot(n, O3n, label='y = $3^n$', color='green', linewidth=2)
plt.plot(n, O5n, label='y = $5^n$', color='yellow', linewidth=2)
plt.plot(n, Ofactn, label='y = $n!$', color='red', linewidth=2)
plt.title('Comparación de orden de crecimiento O($a^n$) vs O($n!$)')
plt.xlabel('Tamaño de la entrada n')
plt.ylabel('Orden de crecimiento')
plt.grid(True)
plt.legend()
plt.show()
```

Observaciones del gráfico: Para esta comparación se decidió restringir los tamaños de entrada para graficar a un máximo de 10, ya que como se puede observar en el gráfico, el orden exponencial y el factorial aumentan de manera excesiva. En el caso del crecimiento exponencial, vemos que el valor de $a$ juega un papel muy importante, pues para valores como $a=2$ y $a=3$ queda muy por debajo del crecimiento factorial; sin embargo, para un valor $a=5$, los papeles se invierten y el crecimiento factorial queda muy por detrás del exponencial. De igual manera, vemos que en tamaños de entrada relativamente pequeños estos órdenes pueden alcanzar escalas que, por ejemplo, el crecimiento cúbico alcanza en valores de $n=150$ aproximadamente. Esto nos da una idea de la magnitud del incremento en el tiempo de ejecución que se puede producir bajo estos órdenes de crecimiento con un cambio tan pequeño en el tamaño de la entrada.

### **$O$($n!$) vs $O$($n^n$)**

Por último, se realiza la comparación entre el crecimiento factorial y el crecimiento $n^n$, denominado superexponencial. De nueva cuenta, como se indica en [2] estamos ante dos órdenes de muy alto costo en los que el más ligero cambio en el tamaño de la entrada produce una variación exagerada en el tiempo de ejecución del algoritmo, por lo que se determina limitar los tamaños de entrada a graficar desde el 1 hasta el 10.

```python
#Creamos los datos a graficar estableciendo los valores de tamaño de entrada y definiendo las funciones dentro de las listas.
n = list(range(1,11))
Ofactn = [math.factorial(i) for i in n]
Onn = [i**i for i in n]

#Graficamos los resultados obtenidos.
plt.plot(n, Ofactn, label='y = $n!$', color='blue', linewidth=2)
plt.plot(n, Onn, label='y = $n^n$', color='red', linewidth=2)
plt.title('Comparación de orden de crecimiento O($n!$) vs O($n^n$)')
plt.xlabel('Tamaño de la entrada n')
plt.ylabel('Orden de crecimiento')
plt.grid(True)
plt.legend()
plt.show()
```

Observaciones del gráfico: Por segunda vez se limitan los tamaños de entrada para graficar a un máximo de 10, ya que, desde el gráfico anterior, se visualizó que el crecimiento factorial aumenta aceleradamente, aunque en esta ocasión se compara con el orden de crecimiento $n^n$, el cual toma escalas aun grandes que hace parecer al factorial como si fuera constante.

# **Tabla comparativa de tiempos de ejecución**

En esta sección del reporte se mostrarán tiempos simulados para los siguientes órdenes de crecimiento: $O(1)$, $O(logn)$, $O(\sqrt{n})$, $O(n^2)$, $O(2^n)$, $O(n!)$, $O(n^n)$ y $O(n^{n^n})$,  con tamaños de entrada $n=10$, $n=100$, $n=1000$ y $n=10000$. Para tal propósito, se utilizará el supuesto de que cada operación tiene un costo de 1 nanosegundo.

Dadas las comparaciones gráficas hechas en la sección anterior, pondremos un manejo de excepciones para aquellos órdenes de crecimiento en los que vimos que las variaciones en el tamaño de la entrada producen cambios desmedidos en los tiempos de ejecución. En caso de que el número a calcular sea exageradamente grande, se indicará como un tiempo de ejecución «infinito». A pesar del uso de manejo de excepciones, para el orden $O(n^{n^n})$, vemos que ni siquiera es posible calcular el valor contemplando un entrada de tamaño 10, pues el resultado sería un 1 seguido por diez mil millones de ceros, por lo que en este caso todos los resultados serán marcados como «infinito».

```python
#Definimos los tamaños de entrada propuestos y nuestras listas de resultados.
n = [10, 100, 1000, 10000]
O1=[]
Ologn = []
Osqrtn=[]
On2=[]
O2n=[]
Ofactn=[]
Onn=[]
Onnn=[]

#Calculamos los tiempos de ejecución simulados para aquellos órdenes moderados.
O1 = [1 for i in n] #Para el orden constante
Ologn = [math.log2(i) for i in n] #Para el orden logn
Osqrtn = [math.sqrt(i) for i in n] #Para el orden sqrt(n)

#Intentamos calcular los tiempos de ejecución simulados para los órdenes de crecimiento más excesivos.
for i in n: #Para el orden On^2
  try:
    On2.append(f"{i**2:.3e}")
  except:
    On2.append("Infinito")

for i in n: #Para el orden O2^n
  try:
    O2n.append(f"{2**i:.3e}")
  except:
    O2n.append("Infinito")

for i in n: #Para el orden n!
  try:
    Ofactn.append(f"{math.factorial(i):.3e}")
  except:
    Ofactn.append("Infinito")

for i in n: #Para el orden n^n
  try:
    Onn.append(f"{i**i:.3e}")
  except:
    Onn.append("Infinito")

Onnn = ["Infinito" for i in n] #Para el orden n^n^n
```

Ya calculados los resultados, procedemos a hacer la tabla comparativa:

```python
df = pd.DataFrame({
    "n": n,
    "O(1)": O1,
    "O(log_2(n))": Ologn,
    "O(sqrt)": Osqrtn,
    "O(n^2)": On2,
    "O(2^n)": O2n,
    "O(n!)": Ofactn,
    "O(n^n)": Onn,
    "O(n^(n^n))": Onnn
})
print(df)
```

Observaciones de la tabla comparativa: En los resultados de la tabla podemos confirmar que los órdenes de crecimiento más moderados son el logarítmico, el de la raíz cuadrada y el cuadrático, ya que para estos órdenes nos fue posible computar el tiempo de ejecución para tamaños de entradas de hasta 10000, manteniéndose tiempos relativamente bajos. Naturalmente el orden constante también, ya que no sufre ninguna alteración independientemente del tamaño de la entrada. Por otro lado, tenemos a los órdenes de crecimiento exponenciales, factoriales y superexponenciales, en los cuales los tiempos de ejecución crecen a tal magnitud que, incluso bajo el supuesto de que una operación tiene un costo de un nanosegundo, se calculan cantidades impronunciables de operaciones, lo que se traduce a miles de millones de segundos de tiempo de ejecución.
